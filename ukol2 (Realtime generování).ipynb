```python
import json
import random
import time
from datetime import datetime
from azure.eventhub import EventHubProducerClient, EventData

CONNECTION_STR = "******"
EVENT_HUB_NAME = "szcasestudykotyk"
```

```python

producer = EventHubProducerClient.from_connection_string(
    conn_str=CONNECTION_STR,
    eventhub_name=EVENT_HUB_NAME
)

def generate_event():
    return {
        "cas_vytvoreni": datetime.utcnow().isoformat(),
        "id_hry": random.choice(["Hra_1", "Hra_2", "Hra_3"]),
        "vyse_sazky": round(random.uniform(10, 500), 2),
        "typ": "vstup"
    }

print("simulace eventů")

try:
    for i in range(900):  #1 event každou sekundu max 15 min
        event = generate_event()
        producer.send_batch([EventData(json.dumps(event))])
        print(f"Odesláno: {event}")
        time.sleep(1)  #každou sekundu 1
finally:
    producer.close()
    print("simulace ukončena po 15min")
```
